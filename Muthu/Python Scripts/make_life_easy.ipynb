```python
import os

path = "../Graphics&Sprites/images_out"
files = os.listdir(path)

files
```

```python
# Get only hex files
hex_files = [x for x in files if x.endswith(".hex")]

hex_files.sort()

hex_files
        
```

```python
# Concatenate all the hex files and write to a new file. Store offset
offset = 0
offsets = []
with open("all_hex.hex", "w") as f:
    for file in hex_files:
        with open(os.path.join(path, file), "r") as f2:
            data = f2.read()
            f.write(data)
            offsets.append(offset)
            # offset is number of lines in the file
            offset += len(data.split("\n")) - 1

offset
```
